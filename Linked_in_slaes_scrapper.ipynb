```python
import requests
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from selenium import webdriver
```

```python
driver=webdriver.Chrome(executable_path="C:\Driver\chromedriver_win32 (3)\chromedriver.exe")
driver.get("https://www.acquisitions.global/members")
driver.maximize_window()
```

```python
from selenium.webdriver.common.keys import Keys
def scroll_down():
    # Scroll down to the bottom of the page
    driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END)
    time.sleep(2)  # Adjust the sleep duration as needed

# Scroll down repeatedly until reaching the bottom of the page
while True:
    # Scroll down
    scroll_down()
    
    # Check if the bottom of the page has been reached
    page_height = driver.execute_script("return document.body.scrollHeight")
    window_height = driver.execute_script("return window.innerHeight")
    if page_height == window_height:
        break

# Close the WebDriver
driver.quit()
```

```python
bx=driver.find_element(By.CSS_SELECTOR,"#content-region > div > div.list-region > div > ul")
```

```python
box=bx.find_elements(By.TAG_NAME,"li")
```

```python
len(box)
```

```python
import pandas as pd
Dic= {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["Full name","First Name","Last Name",'Industry','Title','About','Location','Intro','personal web','Linked in'])
```

```python

for z in range(633,len(box)):
    industry=''
    try:
        location=box[z].find_element(By.CSS_SELECTOR,'[class="member-list-item-location text-color-lighter"]').text
    except:
        location="NA"
    title=box[z].find_element(By.CSS_SELECTOR,'[class="mighty-attribution-badge theme-color-link-text-link navigate"]').text
    box[z].find_element(By.TAG_NAME,"a").click()
    time.sleep(10)
    try:
        name=driver.find_element(By.TAG_NAME,"h1").text
    except:
        continue
    first_name=name.split()[0]
    last_name=name.split()[-1]
    try:
        first_ele=driver.find_element(By.CSS_SELECTOR,'#community-member-about > div:nth-child(3) > div > div.title').text
    except:
        first_ele="NA"

    if first_ele=="Industry":
        industry=driver.find_element(By.CSS_SELECTOR,'#community-member-about > div:nth-child(3) > div > div.content > a').text
    try:
        snd_ele=driver.find_element(By.CSS_SELECTOR,'#community-member-about > div:nth-child(2) > div > div.title').text
    except:
        snd_ele="NA"
    
    if snd_ele=="Industry":
        industry=driver.find_element(By.CSS_SELECTOR,'#community-member-about > div:nth-child(2) > div > div.content > a').text
    
    try:
        about=driver.find_element(By.CSS_SELECTOR,"#community-member-about > div:nth-child(1) > div > div.content").text
    except:
        about="NA"
    try:
        intro=driver.find_element(By.CSS_SELECTOR,'[class="description text-color-body"]').text
    except:
        intro="NA"
    try:
        per_web=driver.find_element(By.CSS_SELECTOR,'[id="personal-links-region"]').find_element(By.TAG_NAME,"a").get_attribute("href")
    except:
        per_web="NA"
    try:
        linked=driver.find_element(By.CSS_SELECTOR,'[data-original-title="Linkedin"]').get_attribute("href")
    except:
        linked="NA"
    if per_web==linked:
        per_web="NA"
    Dic={"Full name":name,"First Name":first_name,"Last Name":last_name,'Industry':industry,'Title':title,'About':about,'Location':location,'Intro':intro,'personal web':per_web,'Linked in':linked}
    gc= gc.append(Dic, ignore_index=True)
    driver.find_element(By.CSS_SELECTOR,'[data-title="Close"]').click()
    time.sleep(2)
    print(gc)
```

```python
gc.to_excel("Members.xlsx")
```
